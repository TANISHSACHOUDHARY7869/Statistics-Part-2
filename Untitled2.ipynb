15) Implement hypothesis testing using Z-statistics for a sample dataset?

Ans:-

```python
import numpy as np
from scipy.stats import norm

def z_test(sample, pop_mean, pop_std, alpha=0.05):
    """
    Perform a two-tailed Z-test.

    Parameters:
    - sample: array-like, sample data
    - pop_mean: float, population mean under null hypothesis
    - pop_std: float, population standard deviation (known)
    - alpha: significance level

    Returns:
    - z_stat: computed Z-statistic
    - p_value: two-tailed p-value
    - result: whether to reject or fail to reject the null hypothesis
    """
    n = len(sample)
    sample_mean = np.mean(sample)

    # Calculate standard error
    se = pop_std / np.sqrt(n)

    # Calculate Z-statistic
    z_stat = (sample_mean - pop_mean) / se

    # Two-tailed p-value
    p_value = 2 * (1 - norm.cdf(abs(z_stat)))

    # Decision
    if p_value < alpha:
        result = "Reject the null hypothesis"
    else:
        result = "Fail to reject the null hypothesis"

    print(f"Sample Mean: {sample_mean:.4f}")
    print(f"Population Mean (H0): {pop_mean}")
    print(f"Population Std Dev: {pop_std}")
    print(f"Sample Size: {n}")
    print(f"Z-statistic: {z_stat:.4f}")
    print(f"P-value: {p_value:.4f}")
    print(f"Significance Level: {alpha}")
    print(f"Result: {result}")

    return z_stat, p_value, result

# Example usage
sample_data = [102, 99, 101, 100, 98, 105, 97, 103, 100, 102]
population_mean = 100
population_std_dev = 5
significance_level = 0.05

z_test(sample_data, population_mean, population_std_dev, significance_level)
```

16) Create a confidence interval for a dataset using Python and interpret the result?

Ans:-

```python
import numpy as np
from scipy import stats

def confidence_interval(data, confidence=0.95):
    """
    Calculate the confidence interval for the mean of the dataset.

    Parameters:
    - data: list or array-like dataset
    - confidence: confidence level (default 0.95)

    Returns:
    - (lower_bound, upper_bound): tuple representing the confidence interval
    """
    data = np.array(data)
    n = len(data)
    mean = np.mean(data)
    std_err = stats.sem(data)  # standard error of the mean

    # t critical value for two-tailed test
    t_crit = stats.t.ppf((1 + confidence) / 2, df=n-1)

    margin_of_error = t_crit * std_err
    lower_bound = mean - margin_of_error
    upper_bound = mean + margin_of_error

    print(f"Sample Mean = {mean:.4f}")
    print(f"Sample Size = {n}")
    print(f"Confidence Level = {confidence*100}%")
    print(f"Margin of Error = {margin_of_error:.4f}")
    print(f"Confidence Interval = [{lower_bound:.4f}, {upper_bound:.4f}]")
    print(f"Interpretation: We are {confidence*100}% confident that the true population mean lies within this interval.")

    return lower_bound, upper_bound

# Example usage:
dataset = [10, 12, 9, 15, 14, 8, 11, 10, 13]
confidence_interval(dataset)
```

```python
import numpy as np
from scipy.stats import norm

def z_test(sample, pop_mean, pop_std, alpha=0.05):
    """
    Perform a one-sample Z-test to compare sample mean to population mean.

    Parameters:
    - sample: array-like, sample data
    - pop_mean: float, known population mean (null hypothesis)
    - pop_std: float, known population standard deviation
    - alpha: significance level (default 0.05)

    Returns:
    - z_stat: computed Z-statistic
    - p_value: two-tailed p-value
    - conclusion: string interpreting the test result
    """
    n = len(sample)
    sample_mean = np.mean(sample)
    se = pop_std / np.sqrt(n)  # standard error

    # Calculate Z statistic
    z_stat = (sample_mean - pop_mean) / se

    # Two-tailed p-value
    p_value = 2 * (1 - norm.cdf(abs(z_stat)))

    # Interpretation
    if p_value < alpha:
        conclusion = "Reject the null hypothesis: sample mean is significantly different from population mean."
    else:
        conclusion = "Fail to reject the null hypothesis: no significant difference between sample mean and population mean."

    print(f"Sample Mean = {sample_mean:.4f}")
    print(f"Population Mean (H0) = {pop_mean}")
    print(f"Population Std Dev = {pop_std}")
    print(f"Sample Size = {n}")
    print(f"Z-statistic = {z_stat:.4f}")
    print(f"P-value = {p_value:.4f}")
    print(f"Significance Level = {alpha}")
    print(f"Conclusion: {conclusion}")

    return z_stat, p_value, conclusion

# Example usage
sample_data = [102, 99, 101, 100, 98, 105, 97, 103, 100, 102]
population_mean = 100
population_std_dev = 5
significance_level = 0.05

z_test(sample_data, population_mean, population_std_dev, significance_level)
```

4) Perform a two-tailed Z-test using Python and visualize the decision region on a plot?

Ans:-

```python
import numpy as np
from scipy.stats import norm
import matplotlib.pyplot as plt

def two_tailed_z_test(sample, pop_mean, pop_std, alpha=0.05):
    n = len(sample)
    sample_mean = np.mean(sample)
    se = pop_std / np.sqrt(n)

    # Z-test statistic
    z_stat = (sample_mean - pop_mean) / se

    # Two-tailed P-value
    p_value = 2 * (1 - norm.cdf(abs(z_stat)))

    # Critical Z-values
    z_critical = norm.ppf(1 - alpha / 2)

    # Decision
    decision = "Reject H0" if abs(z_stat) > z_critical else "Fail to reject H0"

    # Print results
    print(f"Sample Mean: {sample_mean:.4f}")
    print(f"Z-statistic: {z_stat:.4f}")
    print(f"P-value: {p_value:.4f}")
    print(f"Z-critical (Â±): {z_critical:.4f}")
    print(f"Conclusion: {decision}")

    # Visualization
    x = np.linspace(-4, 4, 1000)
    y = norm.pdf(x)

    plt.figure(figsize=(10, 5))
    plt.plot(x, y, label='Standard Normal Distribution')
    plt.fill_between(x, y, where=(x <= -z_critical), color='red', alpha=0.4, label='Rejection Region (left)')
    plt.fill_between(x, y, where=(x >= z_critical), color='red', alpha=0.4, label='Rejection Region (right)')
    plt.axvline(z_stat, color='blue', linestyle='--', label=f'Z = {z_stat:.2f}')
    plt.axvline(-z_critical, color='black', linestyle='--', label=f'-Z critical = {-z_critical:.2f}')
    plt.axvline(z_critical, color='black', linestyle='--', label=f'+Z critical = {z_critical:.2f}')
    plt.title('Two-Tailed Z-Test Visualization')
    plt.xlabel('Z')
    plt.ylabel('Probability Density')
    plt.legend()
    plt.grid(True)
    plt.show()

    return z_stat, p_value, decision

# Example usage:
np.random.seed(42)
sample_data = np.random.normal(loc=102, scale=5, size=30)
population_mean = 100
population_std = 5
alpha = 0.05

two_tailed_z_test(sample_data, population_mean, population_std, alpha)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def visualize_type1_type2_errors(mu0=100, mu1=105, sigma=10, n=30, alpha=0.05):
    """
    Visualize Type I and Type II errors for a right-tailed hypothesis test.

    Parameters:
    - mu0: Mean under the null hypothesis (H0)
    - mu1: Mean under the alternative hypothesis (H1)
    - sigma: Known population standard deviation
    - n: Sample size
    - alpha: Significance level (default is 0.05)
    """

    # Standard error
    se = sigma / np.sqrt(n)

    # Critical value (right-tailed test)
    z_alpha = norm.ppf(1 - alpha)
    critical_value = mu0 + z_alpha * se

    # Beta (Type II error): P(fail to reject H0 | H1 true)
    z_beta = (critical_value - mu1) / se
    beta = norm.cdf(z_beta)

    # X values for plotting
    x = np.linspace(mu0 - 4 * se, mu1 + 4 * se, 1000)

    # PDFs under H0 and H1
    y0 = norm.pdf(x, loc=mu0, scale=se)
    y1 = norm.pdf(x, loc=mu1, scale=se)

    # Plotting
    plt.figure(figsize=(10, 6))
    plt.plot(x, y0, label='Hâ‚€: Î¼ = {}'.format(mu0), color='blue')
    plt.plot(x, y1, label='Hâ‚: Î¼ = {}'.format(mu1), color='green')

    # Shade Type I error region (alpha)
    plt.fill_between(x, y0, where=(x >= critical_value), color='red', alpha=0.4, label='Type I Error (Î±)')

    # Shade Type II error region (beta)
    plt.fill_between(x, y1, where=(x < critical_value), color='orange', alpha=0.4, label='Type II Error (Î²)')

    # Critical value line
    plt.axvline(critical_value, color='black', linestyle='--', label=f'Critical value = {critical_value:.2f}')

    plt.title('Type I and Type II Errors in Hypothesis Testing')
    plt.xlabel('Sample Mean')
    plt.ylabel('Probability Density')
    plt.legend()
    plt.grid(True)
    plt.show()

    # Print results
    print(f"Critical Value (right-tailed): {critical_value:.2f}")
    print(f"Z(Î±) = {z_alpha:.2f}")
    print(f"Z(Î²) = {z_beta:.2f}")
    print(f"Type II Error Probability (Î²): {beta:.4f}")
    print(f"Power of the test (1 - Î²): {1 - beta:.4f}")

# Example usage
visualize_type1_type2_errors(mu0=100, mu1=105, sigma=10, n=30, alpha=0.05)
```

6) Write a Python program to perform an independent T-test and interpret the results?

Ans:-

```python
import numpy as np
from scipy.stats import ttest_ind

def perform_independent_t_test(sample1, sample2, alpha=0.05):
    """
    Perform an independent two-sample T-test and interpret the result.

    Parameters:
    - sample1, sample2: array-like, two independent samples
    - alpha: significance level (default = 0.05)

    Returns:
    - t_stat: T-statistic
    - p_value: P-value of the test
    - conclusion: test decision
    """
    t_stat, p_value = ttest_ind(sample1, sample2, equal_var=False)  # Welch's t-test

    # Interpretation
    if p_value < alpha:
        conclusion = f"Reject the null hypothesis (Î± = {alpha}): Means are significantly different."
    else:
        conclusion = f"Fail to reject the null hypothesis (Î± = {alpha}): No significant difference in means."

    # Print summary
    print("=== Independent T-Test Result ===")
    print(f"Sample 1 Mean = {np.mean(sample1):.2f}")
    print(f"Sample 2 Mean = {np.mean(sample2):.2f}")
    print(f"T-statistic = {t_stat:.4f}")
    print(f"P-value = {p_value:.4f}")
    print(f"Conclusion: {conclusion}")

    return t_stat, p_value, conclusion

# Example usage
np.random.seed(42)
sample1 = np.random.normal(loc=100, scale=10, size=30)
sample2 = np.random.normal(loc=105, scale=10, size=30)

perform_independent_t_test(sample1, sample2)
```

7) Perform a paired sample T-test using Python and visualize the comparison results?

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import ttest_rel

def paired_t_test_with_plot(before, after, alpha=0.05):
    """
    Perform a paired sample T-test and visualize the results.

    Parameters:
    - before, after: array-like, paired data (e.g., before and after treatment)
    - alpha: significance level
    """

    # Perform paired T-test
    t_stat, p_value = ttest_rel(before, after)

    # Calculate mean difference
    differences = np.array(after) - np.array(before)
    mean_diff = np.mean(differences)

    # Interpretation
    if p_value < alpha:
        conclusion = f"Reject Hâ‚€ (Î± = {alpha}): Significant difference between paired samples."
    else:
        conclusion = f"Fail to reject Hâ‚€ (Î± = {alpha}): No significant difference between paired samples."

    # Print results
    print("=== Paired Sample T-Test ===")
    print(f"Mean (Before) = {np.mean(before):.2f}")
    print(f"Mean (After)  = {np.mean(after):.2f}")
    print(f"Mean Difference = {mean_diff:.2f}")
    print(f"T-statistic = {t_stat:.4f}")
    print(f"P-value = {p_value:.4f}")
    print(f"Conclusion: {conclusion}")

    # Visualization
    plt.figure(figsize=(12, 5))

    # Boxplot of Before and After
    plt.subplot(1, 2, 1)
    plt.boxplot([before, after], labels=["Before", "After"])
    plt.title("Paired Sample Distributions")
    plt.ylabel("Value")
    plt.grid(True)

    # Histogram of differences
    plt.subplot(1, 2, 2)
    plt.hist(differences, bins=10, color='lightgreen', edgecolor='black')
    plt.axvline(0, color='red', linestyle='--')
    plt.title("Histogram of Differences (After - Before)")
    plt.xlabel("Difference")
    plt.ylabel("Frequency")
    plt.grid(True)

    plt.tight_layout()
    plt.show()

    return t_stat, p_value, conclusion

# Simulate paired data
np.random.seed(42)
before = np.random.normal(loc=70, scale=5, size=30)
after = before + np.random.normal(loc=2, scale=3, size=30)  # small treatment effect

# Run paired t-test and plot
paired_t_test_with_plot(before, after)
```

8) Simulate data and perform both Z-test and T-test, then compare the results using Python?

Ans:-

```python
import numpy as np
from scipy.stats import norm, t

def compare_z_t_test(pop_mean=100, pop_std=15, sample_size=30, alpha=0.05):
    # Simulate data from normal distribution
    np.random.seed(42)
    sample = np.random.normal(loc=103, scale=pop_std, size=sample_size)
    sample_mean = np.mean(sample)
    sample_std = np.std(sample, ddof=1)
    se = sample_std / np.sqrt(sample_size)

    # Z-test (assumes known population std)
    se_z = pop_std / np.sqrt(sample_size)
    z_stat = (sample_mean - pop_mean) / se_z
    p_value_z = 2 * (1 - norm.cdf(abs(z_stat)))

    # T-test (uses sample std)
    t_stat = (sample_mean - pop_mean) / se
    df = sample_size - 1
    p_value_t = 2 * (1 - t.cdf(abs(t_stat), df))

    # Print comparison
    print("=== Z-Test (Known Population Std) ===")
    print(f"Z-statistic: {z_stat:.4f}")
    print(f"P-value:     {p_value_z:.4f}")

    print("\n=== T-Test (Unknown Population Std) ===")
    print(f"T-statistic: {t_stat:.4f}")
    print(f"P-value:     {p_value_t:.4f}")
    print(f"Degrees of Freedom: {df}")

    print("\n=== Interpretation ===")
    if p_value_z < alpha:
        print(f"Z-test: Reject Hâ‚€ (Î± = {alpha})")
    else:
        print(f"Z-test: Fail to reject Hâ‚€ (Î± = {alpha})")

    if p_value_t < alpha:
        print(f"T-test: Reject Hâ‚€ (Î± = {alpha})")
    else:
        print(f"T-test: Fail to reject Hâ‚€ (Î± = {alpha})")

# Example usage
compare_z_t_test()
```

9) Write a Python function to calculate the confidence interval for a sample mean and explain its significance?

Ans:-

```python
import numpy as np
from scipy.stats import t

def confidence_interval(data, confidence=0.95):
    """
    Calculate the confidence interval for the sample mean.

    Parameters:
    - data: list or array-like, sample data
    - confidence: confidence level (default = 0.95 for 95% confidence)

    Returns:
    - (lower_bound, upper_bound): tuple representing the confidence interval
    """
    data = np.array(data)
    n = len(data)
    mean = np.mean(data)
    std_err = np.std(data, ddof=1) / np.sqrt(n)  # Standard error
    df = n - 1  # Degrees of freedom

    # T critical value
    t_crit = t.ppf((1 + confidence) / 2, df)

    # Margin of error
    margin = t_crit * std_err

    lower_bound = mean - margin
    upper_bound = mean + margin

    print(f"Sample Mean = {mean:.4f}")
    print(f"Confidence Level = {confidence*100:.1f}%")
    print(f"Confidence Interval = ({lower_bound:.4f}, {upper_bound:.4f})")

    return (lower_bound, upper_bound)

# Example usage
np.random.seed(42)
sample_data = np.random.normal(loc=50, scale=10, size=30)
confidence_interval(sample_data, confidence=0.95)
```

```python
import numpy as np
from scipy.stats import t

def calculate_margin_of_error(data, confidence=0.95):
    """
    Calculate the margin of error for a given confidence level using sample data.

    Parameters:
    - data: list or array-like, sample observations
    - confidence: confidence level (default = 0.95 for 95%)

    Returns:
    - margin_of_error: float
    """
    data = np.array(data)
    n = len(data)
    sample_std = np.std(data, ddof=1)           # Sample standard deviation
    se = sample_std / np.sqrt(n)                # Standard error
    df = n - 1                                   # Degrees of freedom
    t_crit = t.ppf((1 + confidence) / 2, df)     # Critical t-value

    margin_of_error = t_crit * se

    print(f"Sample Size (n): {n}")
    print(f"Sample Mean: {np.mean(data):.4f}")
    print(f"Sample Standard Deviation: {sample_std:.4f}")
    print(f"Confidence Level: {confidence*100:.1f}%")
    print(f"Margin of Error: {margin_of_error:.4f}")

    return margin_of_error

# Example usage
np.random.seed(42)
sample_data = np.random.normal(loc=60, scale=12, size=25)
calculate_margin_of_error(sample_data, confidence=0.95)
```

```python
def bayes_theorem(prior, likelihood, false_positive, prior_not=None):
    """
    Compute the posterior probability using Bayes' Theorem.

    Parameters:
    - prior: P(H), prior probability of the hypothesis (e.g., having the disease)
    - likelihood: P(E|H), probability of evidence given hypothesis (e.g., positive test if diseased)
    - false_positive: P(E|Â¬H), probability of evidence given no hypothesis (e.g., positive test if healthy)
    - prior_not: P(Â¬H), optional; otherwise computed as 1 - prior

    Returns:
    - posterior: P(H|E), updated probability of the hypothesis given the evidence
    """

    if prior_not is None:
        prior_not = 1 - prior

    # P(E): total probability of evidence
    evidence = likelihood * prior + false_positive * prior_not

    # Bayes' Theorem
    posterior = (likelihood * prior) / evidence

    print("=== Bayesian Inference ===")
    print(f"Prior (P(H)) = {prior:.4f}")
    print(f"Likelihood (P(E|H)) = {likelihood:.4f}")
    print(f"False Positive Rate (P(E|Â¬H)) = {false_positive:.4f}")
    print(f"Posterior (P(H|E)) = {posterior:.4f}")

    return posterior

# Example: medical diagnosis
prior_disease = 0.01               # 1% prior probability of disease
likelihood_positive = 0.99         # Test is 99% accurate if diseased
false_positive = 0.05              # 5% false positive rate

bayes_theorem(prior_disease, likelihood_positive, false_positive)
```

12) Perform a Chi-square test for independence between two categorical variables in Python?

Ans:-

```python
import pandas as pd
import numpy as np
from scipy.stats import chi2_contingency

# Step 1: Create a contingency table
# Rows = Gender, Columns = Preference
data = np.array([[30, 10],   # Male: 30 like, 10 dislike
                 [20, 20]])  # Female: 20 like, 20 dislike

table = pd.DataFrame(data,
                     index=["Male", "Female"],
                     columns=["Like", "Dislike"])

print("Contingency Table:")
print(table)

# Step 2: Perform Chi-Square Test
chi2, p, dof, expected = chi2_contingency(table)

print("\n=== Chi-Square Test for Independence ===")
print(f"Chi2 Statistic: {chi2:.4f}")
print(f"Degrees of Freedom: {dof}")
print(f"P-value: {p:.4f}")
print("\nExpected Frequencies:")
print(pd.DataFrame(expected, index=["Male", "Female"], columns=["Like", "Dislike"]))

# Step 3: Interpretation
alpha = 0.05
if p < alpha:
    print("\nConclusion: Reject Hâ‚€ â†’ There is a significant association between Gender and Preference.")
else:
    print("\nConclusion: Fail to reject Hâ‚€ â†’ No significant association between Gender and Preference.")
```

13) Write a Python program to calculate the expected frequencies for a Chi-square test based on observed data?

Ans:-

```python
import numpy as np
import pandas as pd

def calculate_expected_frequencies(observed):
    """
    Calculate expected frequencies for a Chi-square test from observed data.

    Parameters:
    - observed: 2D numpy array or pandas DataFrame of observed counts

    Returns:
    - expected: pandas DataFrame of expected frequencies
    """
    if isinstance(observed, pd.DataFrame):
        observed_np = observed.values
    else:
        observed_np = np.array(observed)

    row_totals = observed_np.sum(axis=1, keepdims=True)
    col_totals = observed_np.sum(axis=0, keepdims=True)
    grand_total = observed_np.sum()

    expected_np = (row_totals @ col_totals) / grand_total  # matrix multiplication

    if isinstance(observed, pd.DataFrame):
        expected = pd.DataFrame(expected_np, index=observed.index, columns=observed.columns)
    else:
        expected = pd.DataFrame(expected_np)

    return expected

# Example usage
observed_data = pd.DataFrame(
    [[30, 10],
     [20, 20]],
    index=["Male", "Female"],
    columns=["Like", "Dislike"]
)

print("Observed Frequencies:")
print(observed_data)

expected_freq = calculate_expected_frequencies(observed_data)

print("\nExpected Frequencies:")
print(expected_freq)
```
